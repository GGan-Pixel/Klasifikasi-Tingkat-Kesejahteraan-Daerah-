```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
import joblib

#load dataset
df = pd.read_csv('statistika.csv')

#cek missing values
print(df.isnull().sum())

#kolom numerik

kolom_numerik = [
    'Jumlah_Penduduk', 'Rata_Lama_Sekolah', 'Angka_Partisipasi_Sekolah',
    'Angka_Pengangguran', 'Persentase_Rumah_Layak', 'Akses_Air_Bersih',
    'PDRB_per_Kapita'
]

#isi missing value dengan median
for col in kolom_numerik:
    df[col].fillna(df[col].median(), inplace=True)

#encoding kolom kategorikal
le_provinsi = LabelEncoder()
le_kabkota = LabelEncoder()
le_kategori = LabelEncoder()

df['Provinsi']          = le_provinsi.fit_transform(df['Provinsi'])
df['Kabupaten_Kota']    = le_kabkota.fit_transform(df['Kabupaten_Kota'])
df['Kategori_Wilayah']  = le_kategori.fit_transform(df['Kategori_Wilayah'])

#normalisasi fitur numerik
scaler = MinMaxScaler()
df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])

#simpan scaler dan label encoder untuk digunakan saat prediksi
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(le_provinsi, 'le_provinsi.pkl')
joblib.dump(le_kabkota, 'le_kabkota.pkl')
joblib.dump(le_kategori, 'le_kategori.pkl')

#pisahkan data
x = df.drop('Tingkat_Kesejahteraan', axis=1)
y = df['Tingkat_Kesejahteraan']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

print(f"x_train: {x_train.shape}, y_train: {y_test.shape}")

print(f"Distribusi y_train:\n{y_train.value_counts()}")
```
